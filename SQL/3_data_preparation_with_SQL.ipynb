# Where: Filters

```python
%%sql
SELECT *
FROM CompanyBySector
WHERE 
    Sector = 'Information Technology' 
    AND 
    StockPrice > 1000 
    
```

# HAVING: Filtering on the grouping results

```python
%%sql
SELECT
    Sector, AVG(StockPrice) AS Avg_StockPrice
FROM
    CompanyBySector
GROUP BY Sector
HAVING AVG(StockPrice) > 100
```

# LIKE operator

Wildcard characters in LIKE:

% --> any string

_ --> any single characters

```python
%%sql
SELECT *
FROM CompanyBySector
WHERE CompanyName LIKE '% inc%'
LIMIT 5
```

```python
%%sql
SELECT *
FROM CompanyBySector
WHERE CompanyName NOT LIKE '% inc%'
LIMIT 5
```

```python
%%sql
SELECT *
FROM CompanyBySector
WHERE CompanyName LIKE 'b%'
LIMIT 5
```

```python
%%sql
SELECT *
FROM CompanyBySector
WHERE Ticker LIKE 'GOO_'
```

# LIKE operator with ESCAPE clause

Sometimes the pattern, which you want to match, contains wildcard character e.g., 10%, _20, etc. In this case, you can use the ESCAPE clause to specify the escape character so that MySQL will interpret the wildcard character as a literal character. If you donâ€™t specify the escape character explicitly, the backslash character \ is the default escape character.

For example, if you want to find products whose product codes contain the string _20 , you can use the pattern %\_20% 

# Subqueries in: SELECT, FROM, and WHERE

Select

```python
%%sql

SELECT
    t1.CompanyName,
    t1.Sector,
    t1.StockPrice,
    (SELECT AVG(t2.StockPrice) FROM CompanyBySector t2 WHERE t1.sector = t2.sector) AS AvgStockPrice
FROM
    CompanyBySector t1
```

FROM : AvgStockPrice group by sector where stockPrice > 100

```python
%%sql
SELECT
    t1.Sector,
    AVG(t1.StockPrice) AS AvgStockPrice
FROM
    (SELECT Sector, StockPrice FROM CompanyBySector WHERE StockPrice > 100) t1
GROUP BY t1.Sector
```

WHERE: Find sector where stockPrie is the highest

```python
%%sql
SELECT
    t2.Sector
FROM
    CompanyBySector t2
WHERE
    t2.StockPrice = (SELECT MAX(t1.StockPrice) FROM CompanyBySector t1)
```

# Creating a VIEW (VIRTUAL TABLE)

```python
%%sql
CREATE VIEW Inc_company
AS
SELECT 
    *
FROM 
    CompanyBySector
WHERE
    CompanyName LIKE '% Inc.%'
```

```python
%sql SELECT * FROM Inc_company LIMIT 5
```

# Partitioning with OVER and PARTITION BY: Creating subsets of data and performing calculations on it

```python
%%sql
SELECT
    CompanyName, StockPrice, Sector, AVG(StockPrice) OVER w AS AvgStockPrice
FROM
    CompanyBySector
WINDOW w AS (PARTITION BY Sector)
```
