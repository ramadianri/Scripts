# Conditional Logic: CASE

-- Simple CASE

```python
%%sql
SELECT 
    CompanyName, 
    CASE LENGTH(Ticker)
        WHEN 1 THEN '1 letter'
        WHEN 2 THEN '2 letters'
        WHEN 3 THEN '3 letters'
        ELSE 'more than 3 letters'
    END AS TickerLength
FROM CompanyBySector
LIMIT 5
```

-- Searched CASE

```python
%%sql
SELECT
    CompanyName,
    StockPrice,
    CASE
        WHEN StockPrice < 30 THEN 1
        WHEN (StockPrice >= 30 AND StockPrice <= 100) THEN 2
        ELSE 3
    END AS Tier
FROM CompanyBySector
ORDER BY Tier ASC
```

# Conditional Logic: IF-THEN-ELSE (It can only be used in functions or procedures)

```python
%sql SET @showHighProfile = 1
```

```python
%%sql
IF @showHighProfile = 1 THEN
    BEGIN
        SELECT CompanyName, Sector, StockPrice
        FROM CompanyBySector
        WHERE StockPrice > 100
        ORDER BY StockPrice DESC
    END
ELSE
    BEGIN
        SELECT 'ELSE'
    END
END IF
```

# WHILE-DO loop (It can only be used in functions or procedures)

```python
%%sql

WHILE @MyInteger > 5 DO
    SET @MyInteger = @MyInteger - 1
END WHILE
```
